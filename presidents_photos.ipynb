```python
import tkinter as tk
from tkinter import messagebox
from PIL import Image, ImageTk
import os
import random

# Initialize counters for correct and incorrect answers
correct_answers = 0
incorrect_answers = 0

# Load images from the 'prezydenci' folder
image_folder = 'prezydenci'
image_files = [f for f in os.listdir(image_folder) if f.endswith('.jpg')]

# Shuffle the image files to ensure random order
random.shuffle(image_files)

# Initialize an index to keep track of the current image
current_image_index = 0
current_image_file = ''

# Function to load the next image in the shuffled list
def load_next_image():
    global current_image_index, current_image_file
    if current_image_index < len(image_files):
        current_image_file = image_files[current_image_index]
        image_path = os.path.join(image_folder, current_image_file)
        image = Image.open(image_path)
        image = image.resize((300, 300), Image.ANTIALIAS)
        photo = ImageTk.PhotoImage(image)
        image_label.config(image=photo)
        image_label.image = photo
        current_image_index += 1
    else:
        messagebox.showinfo("Game Over", f"You've gone through all the images!\nCorrect answers: {correct_answers}\nIncorrect answers: {incorrect_answers}")
        root.quit()

# Modify the check_guess function to load the next image on correct guess
def check_guess():
    global current_image_file, correct_answers, incorrect_answers
    user_guess = entry.get().strip()
    if user_guess == current_image_file[:-4]:  # Remove the '.jpg' extension
        correct_answers += 1
        result_label.config(text="Correct!", fg="green")
        submit_button.pack_forget()
        entry.pack_forget()
        result_label.pack()
        next_button.pack()
        show_info()
    else:
        incorrect_answers += 1
        result_label.config(text=f"Incorrect. The correct answer was {current_image_file[:-4]}.", fg="red")
        submit_button.pack_forget()
        next_button.pack()
        entry.pack_forget()
        result_label.pack()
        show_info()
    
    
# create labels that will display the information about the presidents

def show_info():
    global current_image_file
    president = current_image_file[:-4]
    info = database[president]
    info_text = f"Lata sprawowania władzy: {info['years']}\nPrezydent nr: {info['order']}\nPartia: {info['party']}\n\n"
    info_text += "\n".join(info['info'])
    messagebox.showinfo(president, info_text)

# write a function to remove showinfo message box
def remove_info():
    messagebox.showinfo().destroy()


# Create the main window
root = tk.Tk()
root.title("Guess the Picture Game")

# Create and pack the widgets
image_label = tk.Label(root)
image_label.pack()

entry = tk.Entry(root, width=50)
entry.pack()

next_button = tk.Button(root, text="Next", command=lambda: [entry.pack(), next_button.pack_forget(),submit_button.pack(), load_next_image(), result_label.pack_forget()])

submit_button = tk.Button(root, text="Submit", command=lambda: [check_guess(), entry.delete(0, tk.END)])
submit_button.pack()

# Bind the Enter key to the submission
#root.bind('<Return>', lambda event: [check_guess(), entry.delete(0, tk.END)])
submit_button.pack()

result_label = tk.Label(root, text="")
result_label.pack()

# Load the first image
load_next_image()

# Run the main loop
root.mainloop()
```
