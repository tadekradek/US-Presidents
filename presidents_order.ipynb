```python
import tkinter as tk
from tkinter import messagebox

# Define the correct order of elements
presidents = [
    "George Washington",
    "John Adams",
    "Thomas Jefferson",
    "James Madison",
    "James Monroe",
    "John Quincy Adams",
    "Andrew Jackson",
    "Martin Van Buren",
    "William Henry Harrison",
    "John Tyler",
    "James Polk",
    "Zachary Taylor",
    "Millard Fillmore",
    "Franklin Pierce",
    "James Buchanan",
    "Abraham Lincoln",
    "Andrew Johnson",
    "Ulysses Grant",
    "Rutherford Hayes",
    "James Garfield",
    "Chester Arthur",
    "Grover Cleveland",
    "Benjamin Harrison",
    "Grover Cleveland", 
    "William McKinley",
    "Theodore Roosevelt",
    "William Howard Taft",
    "Woodrow Wilson",
    "Warren Harding",
    "Calvin Coolidge",
    "Herbert Hoover",
    "Franklin Roosevelt",
    "Harry Truman",
    "Dwight Eisenhower",
    "John Kennedy",
    "Lyndon Johnson",
    "Richard Nixon",
    "Gerald Ford",
    "Jimmy Carter",
    "Ronald Reagan",
    "George Bush Senior",
    "Bill Clinton",
    "George Bush",
    "Barack Obama",
    "Donald Trump",
    "Joe Biden"
]

def check_order():
    user_list = [entry.get() for entry in entries]
    user_list = [item.strip() for item in user_list]

# Create the main window
root = tk.Tk()
root.title("Order the Elements Game")

entries = []
columns = 4  # Number of columns
rows = (len(presidents) + columns - 1) // columns  # Calculate the number of rows needed

for col in range(columns):
    frame = tk.Frame(root)
    frame.pack(side=tk.LEFT, padx=10)
    for row in range(rows):
        index = col * rows + row
        if index < len(presidents):
            entry_frame = tk.Frame(frame)
            entry_frame.pack(pady=5)
            label = tk.Label(entry_frame, text=f"{index + 1}.")
            label.pack(side=tk.LEFT)
            entry = tk.Entry(entry_frame, width=50)
            entry.pack(side=tk.LEFT)
            entries.append(entry)

# Function to check individual entries
def check_individual_entries():
    user_list = [entry.get().strip() for entry in entries]
    
    for i, (user_entry, correct_entry) in enumerate(zip(user_list, presidents)):
        if user_entry == correct_entry:
            labels[i].config(text="✔", fg="green")
        else:
            labels[i].config(text="✘", fg="red")
    
    update_statistics()
    display_correct_order()

# Create labels for individual checks
labels = []
for entry in entries:
    label = tk.Label(entry.master, text="", width=2)
    label.pack(side=tk.LEFT)
    labels.append(label)

# Create the submit button
# Create a frame for the submit button and statistics
bottom_frame = tk.Frame(root)
bottom_frame.pack(side=tk.BOTTOM, pady=10)

# Create the submit button
submit_button = tk.Button(bottom_frame, text="Submit", command=lambda: [check_order(), check_individual_entries()])
submit_button.pack()

# Create a label to display the statistics
stats_label = tk.Label(bottom_frame, text="Statistics: Correct - 0, Incorrect - 0")
stats_label.pack()

# Function to update statistics
def update_statistics():
    correct_count = sum(1 for label in labels if label.cget("text") == "✔")
    incorrect_count = len(labels) - correct_count
    total_count = len(labels)
    percentage = (correct_count / total_count) * 100
    stats_label.config(text=f"Statistics: Correct - {correct_count}, Incorrect - {incorrect_count}, Percentage - {percentage:.2f}%")

def display_correct_order():
    correct_order_label.config(text="\n".join(presidents))

# Create a label to display the correct order
correct_order_label = tk.Label(root, text="", justify=tk.LEFT)
correct_order_label.pack()

result_label = tk.Label(root, text="")
result_label.pack()

# Run the main loop
root.mainloop()
```
